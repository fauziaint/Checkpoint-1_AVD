```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Fastfood Nutrition Data.zip")

df
```

# **VISUALISASI DATA**

**1. Bar Chart**

```python
plt.figure(figsize=(12,6))
plt.bar(df['item'].value_counts().head(10).index, df['item'].value_counts().head(10).values, color='pink')
plt.title('Top 8 Menu dengan Kalori Tertinggi')
plt.xlabel('Menu Makanan')
plt.ylabel('Jumlah Kalori')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='item', data=df, hue='restaurant', palette='pastel', order=df['item'].value_counts().head(10).index)
plt.title('Top 8 Menu dengan Kalori Tertinggi')
plt.xlabel('Brand')
plt.ylabel('Jumlah Kalori')
plt.xticks(rotation=45)
plt.show()
```

**2. Scatter Plot**

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['calories'], df['total_fat'], c='pink', alpha=0.5)
plt.title('Hubungan antara Kalori VS Lemak')
plt.xlabel('Kalori')
plt.ylabel('Lemak (g)')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='total_fat', y='calories', hue='restaurant', data=df)
plt.title('Hubungan antara Kalori VS Lemak')
plt.xlabel('Lemak')
plt.ylabel('Kalori')
plt.legend(title='restaurant')
plt.show()
```

**3. Box Plot**

```python
plt.figure(figsize=(8, 5))
df.boxplot(column='calories', by='restaurant', grid=False, patch_artist=True)
plt.title('Variasi Kandungan Kalori di Berbagai Restoran')
plt.xlabel('Restoran')
plt.ylabel('Kandungan Kalori')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='restaurant', y='calories',hue="restaurant", data=df, palette='pastel')
plt.title('Variasi Kandungan Kalori di Berbagai Restoran')
plt.xlabel('Restoran')
plt.ylabel('Kandungan Kalori')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# **INSIGHT DAN TINDAKAN**

**Kandungan Kalori Rata-Rata di Setiap Restoran**
1. Beberapa restoran punya kalori lebih tinggi, sementara yang lain lebih rendah.
2. Tindakan: Konsumen bisa pilih restoran dengan menu lebih rendah kalori, dan restoran bisa tawarkan opsi rendah kalori.

**Hubungan Kalori vs. Lemak**
1. Makanan tinggi kalori biasanya juga tinggi lemak.
2. Tindakan: Konsumen dapat memilih makanan rendah lemak tetapi tetap mengenyangkan, jika mereka ingin mengurangi konsumsi lemak.

**Variasi Kandungan Kalori di Setiap Restoran**
1. Ada menu dengan kalori ekstrem (outlier).
2. Tindakan: Konsumen dapat menghindari menu yang ekstrem jika mereka ingin menjaga pola makan sehat, sementara restoran bisa menyeimbangkan kandungan gizi.
